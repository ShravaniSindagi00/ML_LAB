Activity to do: <br>

Develop a program to implement k-Nearest Neighbour algorithm to classify the randomly generated 100 values of X in the range of [0 1]. Perform the following based on dataset generated" a. Label the first 50 points {x1, ..., x50} as follows; if xi <= 0.5 then xi blongs to class C1 else class C2. b. Classify the remaining points {x51, ..., x100} using KNN. Perform this for k = 1,2,3,4,5,20,30.

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# 1. Generate 100 random values in [0, 1]
X = np.random.rand(100, 1)

# 2. Label the first 50 points:
y = np.array(['C1' if x <= 0.5 else 'C2' for x in X[:50].flatten()])

# Split the data:
X_train = X[:50]
X_test = X[50:]

# 3. Classify the remaining points using KNN with different k values
k_values = [1, 2, 3, 4, 5, 20, 30]

print("KNN Classification Results:\n")

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y)

    y_pred = knn.predict(X_test)

    print(f"\n--- k = {k} ---")
    for i in range(len(X_test)):
        print(f"x = {X_test[i][0]:.3f} --> Predicted Class: {y_pred[i]}")
```
